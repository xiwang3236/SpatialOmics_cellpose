```python
import pandas as pd
import numpy as np
```

```python
sample_name = "1835od"
centroids_file = f"R:/Cellpose-resegmentation/cellpose_cell_centroids_{sample_name}.csv"
centroids_df = pd.read_csv(centroids_file)
centroids_df
```

```python
print(f"Centroids file shape: {centroids_df.shape}")
print(f"Centroids columns: {centroids_df.columns.tolist()}")
```

```python
expression_file = f"R:/Cellpose-resegmentation/{sample_name}_gene_expression_matrix.csv"
expression_df = pd.read_csv(expression_file)
expression_df
```

```python

# Merge with gene expression data
merged_df = pd.merge(
    
    centroids_df,
    expression_df, 
    left_on="cell_id", 
    right_on="cell_id", 
    how='left'
)
```

```python
merged_df
```

```python
# Get columns after the 4th column
later_columns = merged_df.columns[1:]
# Fill empty values with 0 for all columns after column 4
merged_df[later_columns] = merged_df[later_columns].fillna(0)
```

```python
merged_df
```

```python
num_cols = merged_df.shape[1]

# Delete the last two columns
merged_df = merged_df.iloc[:, :-2]

# Store columns 2 and 3 (index 1 and 2) for moving to the end
# Adjust these indices if needed based on your actual data structure
centroid_x = merged_df.iloc[:, 1].copy()
centroid_y = merged_df.iloc[:, 2].copy()

# Drop the original columns 2 and 3
merged_df_new = merged_df.drop(merged_df.columns[[1, 2]], axis=1)

# Add the stored columns back at the end with new names
merged_df_new['centroid_x'] = centroid_x
merged_df_new['centroid_y'] = centroid_y

```

```python
# Multiply centroid_x and centroid_y by 0.2125 
merged_df_new['centroid_x'] = merged_df_new['centroid_x'] * 0.2125
merged_df_new['centroid_y'] = merged_df_new['centroid_y'] * 0.2125
```

```python
# Check the result
merged_df_new
```

```python
import pandas as pd

# Assuming transcripts_assigned_ture is your DataFrame
output_path = f"R:/Cellpose-resegmentation/gene_expression_matrix_{sample_name}.csv"

# Save the DataFrame as a Parquet file
merged_df_new.to_csv(output_path, index=False)

print(f"DataFrame saved successfully to {output_path}")
```
