```python
# Essential imports
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from shapely.geometry import Point, Polygon
import seaborn as sns  # For better color palettes (optional)

# Optional imports for additional functionality

from pathlib import Path
```

```python
# Read the CSV file into a pandas DataFrame
file_path = r'G:\Alexis_data\Project\SpatialOmics_cellpose\results\segmenatation\1889od\cellpose_cell_boundaries_1889od.csv'
df = pd.read_csv(file_path)
```

```python
def plot_cell_boundaries(df, figsize=(10, 10), title="Cell Boundaries"):
    """
    Plot cell boundaries from vertex coordinates.
    
    Args:
        df: DataFrame with cell_id, vertex_x, vertex_y columns
        figsize: Figure size tuple (width, height)
        title: Plot title
    """
    plt.figure(figsize=figsize)
    
    # Plot boundaries for each cell
    for cell_id in df['cell_id'].unique():
        cell_data = df[df['cell_id'] == cell_id]
        
        # Connect vertices to form the boundary
        plt.plot(cell_data['vertex_x'], 
                cell_data['vertex_y'], 
                '-', 
                linewidth=1, 
                alpha=0.5)
    
    plt.title(title)
    plt.xlabel('X Coordinate')
    plt.ylabel('Y Coordinate')
    plt.axis('equal')
    plt.grid(True, alpha=0.3)
    plt.show()

# Usage:
plot_cell_boundaries(df, title="Scaled Cell Boundaries")
```
